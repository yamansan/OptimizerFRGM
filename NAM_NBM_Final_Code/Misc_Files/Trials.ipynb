```python
import pickle
import matplotlib.pyplot as plt
import numpy as np

# Path to your pickle file
pickle_file_path = 'distributions/percentile_tables.pkl'

# Open the pickle file in read-binary mode
with open(pickle_file_path, 'rb') as file:
    data = pickle.load(file)

# Plot PDF and CDF together for each breakeven level
for breakeven, info in data.items():
    dips = info['dips']
    
    # Sort the dips for CDF calculation
    sorted_dips = np.sort(dips)
    cdf = np.arange(1, len(sorted_dips) + 1) / len(sorted_dips)
    
    # Define bins starting from 6 with a unit width
    min_bin = 6
    max_bin = np.ceil(max(dips)) + 1
    bins = np.arange(min_bin, max_bin, 1)
    
    # Create a new figure for each breakeven level
    plt.figure(figsize=(10, 6))
    
    # Plot the PDF
    plt.hist(dips, bins=bins, density=True, alpha=0.6, color='g', label='PDF')
    
    # Plot the CDF
    plt.plot(sorted_dips, cdf, marker='.', linestyle='-', color='b', label='CDF')
    
    # Add labels and title
    plt.title(f'PDF and CDF of Dips for Breakeven: {breakeven}')
    plt.xlabel('Dip (bps)')
    plt.ylabel('Density / CDF')
    plt.legend()
    
    # Show the plot
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

# --- helpers ----------------------------------------------------------
_TICK32 = 1 / 32        # 0.03125
_HALF32 = 1 / 64        # 0.015625

def decimal_to_zn(px: float) -> str:
    """
    Convert a decimal ZN price (e.g. 111.140625) into 32nds string (111-04+).
    Handles half-ticks.  Rounds to the nearest half-tick.
    """
    pts = int(px)                        # whole points (111)
    frac_dec = px - pts                  # 0  0.999
    
    # express remainder in half-ticks (each = 1/64)
    half_ticks = int(round(frac_dec / _HALF32))     # 0,1,2,,63
    
    ticks32, half = divmod(half_ticks, 2)           # 031  /  0|1
    
    # roll-over in case rounding pushed us to 32-00
    if ticks32 == 32:
        pts += 1
        ticks32 = 0
        half = 0
    
    return f"{pts}-{ticks32:02d}{'+' if half else ''}"

# Matplotlib formatter object
zn_formatter = FuncFormatter(lambda x, pos: decimal_to_zn(x))
# ----------------------------------------------------------------------

# ---------- example plot of df2 ---------------------------------------
fig, ax = plt.subplots(figsize=(15, 10))

ax.plot(df2['Datetime'], df2['Close'], lw=1)  # or whatever column you wish
ax.set_xlabel('Time')
ax.set_ylabel('Price (ZN format)')

# Apply the formatter
ax.yaxis.set_major_formatter(zn_formatter)

# If you want finer Y-grid lines at half-tick increments:
ax.yaxis.set_major_locator(plt.MultipleLocator(_TICK32))      # 1/32 steps
ax.yaxis.set_minor_locator(plt.MultipleLocator(_HALF32))      # 1/64 steps
ax.grid(True, which='both', axis='y', linestyle='--', alpha=0.3)

plt.tight_layout()
plt.show()
```
